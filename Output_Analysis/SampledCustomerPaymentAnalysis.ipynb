```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hive_data = pd.read_csv("../Output/Hive/Complex_Queries/SampledCustomerPaymentAnalysis/000000_0.csv")
```

```python
hive_data
```

```python
# Pivoting the data for plotting
pivot_data = hive_data.pivot_table(
    index=['product_line', 'customer_type'],
    columns='payment',
    values='payment_revenue',
    aggfunc='sum',
    fill_value=0
)

# Plotting the data
pivot_data.plot(kind='bar', stacked=True, figsize=(14, 8))
plt.title('Payment Revenue by Product Line and Customer Type')
plt.xlabel('Product Line, Customer Type')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.legend(title='Payment Method')
plt.show()
```

```python
product_lines = hive_data['product_line'].unique()

# Creating a plot for each product line
for product in product_lines:
    plt.figure(figsize=(10, 6))
    # Filter data for the current product line
    product_data = hive_data[hive_data['product_line'] == product]
    sns.barplot(data=product_data, x='customer_type', y='payment_revenue', hue='payment', palette='Set2')
    plt.title(f'Payment Revenue for {product}')
    plt.xlabel('Customer Type')
    plt.ylabel('Revenue')
    plt.legend(title='Payment Method')
    plt.show()
```
