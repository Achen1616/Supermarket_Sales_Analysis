```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hive_data = pd.read_csv("../Output/Hive/Complex_Queries/TopCityProductSales/000000_0.csv")
```

```python
hive_data
```

```python
# Creating a grouped bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='product_line', y='product_line_revenue', hue='city', data=hive_data, palette='Set2')
plt.title('Product Line Revenue by City and Customer Type')
plt.xlabel('Product Line')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.legend(title='City')
plt.tight_layout()
plt.show()
```
