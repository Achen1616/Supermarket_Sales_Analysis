```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hive_data = pd.read_csv("../Output/Hive/Complex_Queries/MonthlyProductRev/000000_0.csv")
```

```python
hive_data
```

```python
# Unique months in the data
unique_months = hive_data['month'].unique()

# Plotting separate bar graphs for revenue and ratings for each month
for month in unique_months:
    # Filter data for the month
    month_data = hive_data[hive_data['month'] == month]
    
    # Plot for Monthly Revenue
    plt.figure(figsize=(10, 6))
    plt.bar(month_data['product_line'], month_data['monthly_revenue'], color='blue', label='Monthly Revenue')
    plt.xlabel('Product Line')
    plt.ylabel('Monthly Revenue')
    plt.ylim(0, 30000)  # Setting a uniform y-axis for revenue
    plt.title(f'Monthly Revenue for Month {month}')
    plt.xticks(rotation=90)
    plt.legend()
    plt.show()

    # Plot for Average Rating
    plt.figure(figsize=(10, 6))
    plt.bar(month_data['product_line'], month_data['avg_rating'], color='green', label='Avg Rating')
    plt.xlabel('Product Line')
    plt.ylabel('Avg Rating')
    plt.ylim(0, 10)  # Setting a uniform y-axis for ratings
    plt.title(f'Average Ratings for Month {month}')
    plt.xticks(rotation=90)
    plt.legend()
    plt.show()
```
